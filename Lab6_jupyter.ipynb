```python
# Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset from 'homeprices.csv'
df = pd.read_csv('homeprices.csv')

# Display the dataset
print("Dataset:")
print(df)

# Visualize the data using a scatter plot
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Area vs Price')
plt.scatter(df['area'], df['price'], color='blue', marker='o')
plt.show()

# Separate independent (X) and dependent (Y) variables
X = df[['area']]  # Independent variable (area)
Y = df['price']   # Dependent variable (price)

# Create a Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict prices for the given areas in the dataset
predicted_prices = model.predict(X)

# Display predictions alongside actual prices
df['Predicted Price'] = predicted_prices
print("\nActual vs Predicted Prices:")
print(df)

# Visualize the regression line
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression Line')
plt.scatter(df['area'], df['price'], color='blue', marker='o', label='Actual Prices')
plt.plot(df['area'], predicted_prices, color='red', label='Regression Line')
plt.legend()
plt.show()

# Display the regression equation
slope = model.coef_[0]
intercept = model.intercept_
print(f"\nRegression Equation: Y = {slope:.2f}X + {intercept:.2f}")

# Verify the prediction for areas in the dataset
for area in df['area']:
    predicted_price = model.predict([[area]])
    print(f"Predicted price for {area} sq ft: ${predicted_price[0]:.2f}")
```

```python
# Import required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Define the dataset
data = {
    'area': [2600, 3000, 3200, 3600, 4000],
    'price': [550000, 565000, 610000, 680000, 725000]
}
df = pd.DataFrame(data)

# Display the dataset
print("Dataset:")
print(df)

# Visualize the data using a scatter plot
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Area vs Price')
plt.scatter(df['area'], df['price'], color='blue', marker='o')
plt.show()

# Separate independent (X) and dependent (Y) variables
X = df[['area']]  # Independent variable (area)
Y = df['price']   # Dependent variable (price)

# Create a Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict prices for the areas: 5000 sq ft, 8000 sq ft, and 9000 sq ft
areas_to_predict = [[5000], [8000], [9000]]
predicted_prices = model.predict(areas_to_predict)

# Display the predictions
print("\nPredicted prices for specific areas:")
for area, price in zip([5000, 8000, 9000], predicted_prices):
    print(f"Predicted price for {area} sq ft: ${price:,.2f}")

# Visualize the regression line and predictions
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression Line with Predictions')
plt.scatter(df['area'], df['price'], color='blue', marker='o', label='Actual Prices')
plt.plot(df['area'], model.predict(X), color='red', label='Regression Line')

# Plot predicted prices
for area, price in zip([5000, 8000, 9000], predicted_prices):
    plt.scatter(area, price, color='green', marker='x', label=f'Prediction for {area} sq ft')

plt.legend()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('diabetes_data.csv')

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Define independent variables (X) and dependent variable (Y)
# Let's predict 'Outcome' based on other features
X = df[['BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]  # Independent variables
Y = df['Outcome']  # Dependent variable

# Create a Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, Y)

# Predict the outcome for all data points
predictions = model.predict(X)

# Display the predicted values alongside the actual values
df['Predicted Outcome'] = predictions
print("\nActual vs Predicted Outcomes:")
print(df[['Outcome', 'Predicted Outcome']])

# Plotting the predictions vs actual outcomes
plt.scatter(df['Age'], df['Outcome'], color='blue', label='Actual Outcome')
plt.scatter(df['Age'], df['Predicted Outcome'], color='red', label='Predicted Outcome')
plt.xlabel('Age')
plt.ylabel('Outcome')
plt.title('Actual vs Predicted Outcomes')
plt.legend()
plt.show()

# You can also predict the outcome for new data points.
# For example, let's predict the outcome for a new person with specific feature values:
new_data = pd.DataFrame({
    'BloodPressure': [70],
    'SkinThickness': [30],
    'Insulin': [100],
    'BMI': [32],
    'DiabetesPedigreeFunction': [0.45],
    'Age': [40]
})

new_prediction = model.predict(new_data)
print("\nPredicted Outcome for new data:", new_prediction[0])
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('diabetes_data.csv')

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Define independent variables (X) and dependent variable (Y)
X = df[['BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]  # Independent variables
Y = df['Outcome']  # Dependent variable

# Create a Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, Y)

# Predict the outcome for all data points
predictions = model.predict(X)

# Display the predicted values alongside the actual values
df['Predicted Outcome'] = predictions
print("\nActual vs Predicted Outcomes:")
print(df[['Outcome', 'Predicted Outcome']])

# Plotting the predictions vs actual outcomes
plt.scatter(df['Age'], df['Outcome'], color='blue', label='Actual Outcome')
plt.scatter(df['Age'], df['Predicted Outcome'], color='red', label='Predicted Outcome')
plt.xlabel('Age')
plt.ylabel('Outcome')
plt.title('Actual vs Predicted Outcomes')
plt.legend()
plt.show()

# Example: Predicting the outcome for new data points (like predicting price for specific areas)
new_data = pd.DataFrame({
    'BloodPressure': [70, 80, 90],  # Example BloodPressure values
    'SkinThickness': [30, 40, 25],  # Example SkinThickness values
    'Insulin': [100, 200, 50],  # Example Insulin values
    'BMI': [32, 33, 29],  # Example BMI values
    'DiabetesPedigreeFunction': [0.45, 0.67, 0.56],  # Example DiabetesPedigreeFunction values
    'Age': [40, 50, 60]  # Example Age values
})

# Predict the outcome for the new data
new_predictions = model.predict(new_data)
print("\nPredicted Outcomes for new data:")
for idx, prediction in enumerate(new_predictions):
    print(f"Prediction for data point {idx+1}: {prediction:.2f}")
```
